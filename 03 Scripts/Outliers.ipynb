### Find outliers and see their impact on the data

```python
# Find Q1 and Q3 of 'us_dollars_surplus'
df['us_dollars_surplus'].describe()
```

```python
# Create subset of outliers of 'us_dollars_surplus'
# IQR = 1244078 - 18994.06
# outlier > 1.5 x 1225083.94 + 1244078
# outlier > 3081703.91
df_dollars = df[df['us_dollars_surplus'] > 3081703.91]
```

```python
df_dollars.head(50)
```

```python
df_dollars.shape
```

```python
# Find Q1 and Q3 of 'total_mtco2e_footprint'
df['total_mtco2e_footprint'].describe()
```

```python
# IQR = 1702.293 - 34.24223
# outlier > 1.5 x 1668.05077 + 1702.293
# outlier > 4204.369155
df_mtco2e = df[df['total_mtco2e_footprint'] > 4204.369155]
```

```python
df_mtco2e.head(50)
```

```python
df_mtco2e.shape
```

```python
# What percent of the df are outliers from 'us_dollars_surplus'?
(len(df_dollars) / len(df)) * 100
```

```python
# What percent of the df are outliers from 'total_mtco2e_footprint'?
(len(df_mtco2e) / len(df)) * 100
```

```python
# What percent of the surplus in dollar value are the outliers responsible for?
(df_dollars['us_dollars_surplus'].sum() / df['us_dollars_surplus'].sum()) * 100
```

```python
# What percent of the metric ton of carbon emissions are the outliers responsible for?
(df_mtco2e['total_mtco2e_footprint'].sum() / df['total_mtco2e_footprint'].sum()) * 100
```
