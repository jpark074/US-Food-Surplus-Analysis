## 04. Data Consistency Check

### 04.1 Mixed Type Data


```python
# Check or mixed-typ data in df
for col in df_food_surplus.columns.tolist():
    weird = (df_food_surplus[[col]].applymap(type) != df_food_surplus[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_food_surplus[weird]) > 0:
        print(col)
    else:
        print('not mixed')
```
