## 02. Creating subset for correlation

```python
# Create subset without variables not suitable for exploratory visual analysis
sub = df_food_surplus.drop(columns= ['Unnamed: 0', 'year', 'state', 'sector', 'sub_sector', 'sub_sector_category', 'food_type'])
```

```python
# Create correlation matrix
sub.corr()
```

## 03. Correlation matrix heatmap

```python
# Create correlation matrix heatmap
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(25,25))

corr = sns.heatmap(sub.corr(), annot = True, ax = ax) 
```

In general as most of the variables are measurements in tons, if the coefficient indicates a positive relationship then more of that variable, more of the corresponding variable in tons. For example, the coefficient of the relationship between tons_landfilled and upstream_mtco2e coefficient is 0.79, a strong positive correlation. This indicates that more surplus landfilled in tons means more upstream metric tons of C02. Overall, most of the coefficients are positive and the ones that are negative often have extremely weak correlation.

## 07. Revisting questions

Most of the question from the previous task used a qualitative nominal variable, which could not be visualized using the correlation heat map and scatterplots. Therefore, new questions were made for the explanatory visual analysis. The questions were: <br>

**Does more food surplus mean higher impact on the environment?**
* The variable tons_surplus has a moderate positive correlation with total_mtco2e_footprint and gallons_of_water_footprint. However, us_dollars_surplus, the value of the food surplus in us dollars has a strong positive correlation with mtco2e and gallons of water footprint. Furthermore, the surplus of food itself does not have an impact, but the creation, consumption and disposal of food might. This can be seen through the positive correlations between the variables tons_eaten and upstream_mtco2e_footprint, as well as, gallons_of_water_footprint. Additionally, tons_landfill and the two environmental measures. <br>

**Which method of disposing surplus foods create the most metric ton of carbon emissions?**
* Of the disposal methods, landfills have the highest correlation with carbon emissions. <br>

**Does water usage (gallons_water_footprint) have an impact on carbon emissions?**
* The variable gallons_water_footprint has a strong positive correlation with upstream_mtco2e_footprint and total_mtco2e_footprint, but a weak correlation with downstream_mtco2e_footprint. <br>

**What factors create the food surplus and does most of it end up as waste?**
* The variable tons_surplus has high correlation with tons_inedible and tons_not_fit_for_human_consumption. Additionally, there is or nearly is a linear relationship with tons_waste and tons_uneaten.
